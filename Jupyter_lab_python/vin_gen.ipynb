```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
def gen_w(x, sigma=1):
    y = np.zeros_like(x)
    for i, val in enumerate(x[1:]):
        y[i+1] = y[i] + np.random.normal(0, sigma*np.sqrt((val-x[i])))
    return y
```

```python
a = 1
b = 2
N = 41
x = np.linspace(a, b, N)
for i in range(1000):
    plt.plot(x, gen_w(x))
plt.show()
```

```python
a = 1
b = 3
N = 41
x = np.linspace(a, b, N)
for i in range(5000):
    plt.plot(x, gen_w(x))
plt.plot(x, 3*np.sqrt(x-a), '--', color='black', linewidth=3.5)
plt.plot(x, -3*np.sqrt(x-a), '--', color='black', linewidth=3.5)
plt.show()
```

```python
# 4.2
a = 0
b = 15
N = 1001
M = 5000
t1 = 200
t2 = 500
t3 = 741
x = np.linspace(a, b, N)
#print(x[t2])
res = np.array([0]*M, dtype=np.float)
res_2 = np.array([0]*10, dtype=np.float)
for k in range(10):
    for i in range(M):
        xx = gen_w(x)
        res[i] = xx[t1]*xx[t2]*xx[t3]
    res_2[k] = res.mean()
res_2.mean()
```
