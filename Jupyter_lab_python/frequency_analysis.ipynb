```python
#lab1.1
import codecs
warpiece=""
with codecs.open('warpiece.txt', encoding='utf-8') as fin:
    #line = next(fin)
    #print(line.strip())
    for line in fin.readlines():
        warpiece+=line
        warpiece+='\n'
fin.close()
print(warpiece)



check_sample = "абвгдежзийклмнопрстуфхцчшщъыьэюяабвгдежзийклмнопрстуфхцчшщъыьэюя"
#lower = "абвгдежзийклмнопрстуфхцчшщъыьэюяабвгдежзийклмнопрстуфхцчшщъыьэюя"
#upper = "АБВГДЕЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯАБВГДЕЖЗИЙКЛМНОПРСТУФХЦЧШЩЪЫЬЭЮЯ"

encrypt = warpiece
key = int(input())
encrypt = encrypt.lower()
encrypted =""
for i in encrypt:
    if i in check_sample:
        pos = check_sample.find(i)
        newPos = pos+key
        encrypted = encrypted+check_sample[newPos]
    else:
        encrypted = encrypted+i
print('%%%%%%%%%%%%%%%%%%%%%%%%%'*5)
print(encrypted)
```

```python
import collections
test  = collections.Counter(encrypted)
letters_sum = 0
for i in check_sample:
    #print(i, test[i])
    letters_sum+=test[i]
#print(letters_sum)

chapter_frequency = {}
for i in check_sample:
    chapter_frequency[i] = test[i]/letters_sum
chapter_frequency
```

```python
chapter_lst = []
for i in chapter_frequency:
    chapter_lst.append(chapter_frequency[i])
#chapter_lst
basis_lst = []
for i in basis_frequency:
    basis_lst.append(chapter_frequency[i])
#basis_lst
```

```python
srtd_bf = sorted(basis_frequency.items(), key=lambda x: x[1])
srtd_cf = sorted(chapter_frequency.items(), key=lambda x: x[1])
```

```python
ch_to_ch = {}
for i, j in zip(srtd_bf, srtd_cf):
    ch_to_ch[j[0]]=i[0]
ch_to_ch
```

```python
final_txt = ""
for i in encrypted:
    if i not in check_sample:
        final_txt += i
        continue
    final_txt += ch_to_ch[i]

print(final_txt)
```

```python
accuracy = 0
for i, j in zip(warpiece, final_txt):
    if i==j:
        accuracy+=1
print(accuracy/len(warpiece))
```
