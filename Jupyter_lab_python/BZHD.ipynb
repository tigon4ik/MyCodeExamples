```python
import numpy as np
import matplotlib.pyplot as plt

Phi = 10;
n = 2;
Q = 10*np.sqrt(n);
F = 1;
omega = 4*np.pi;
x = (1/5)*n;
y = (1/5)*n;
h = 6;
R = x+y; #
A = (1/2)*h;
B = (1/2)*h;
B1= 0;
c=332
```

```python
freq = [125,250,500,1000,2000,4000,8000]
delta = [0.8, 1.5, 3.8, 12.1, 40, 109, 196]
lamb = [c/i for i in freq]
a = np.sqrt((h-A)**2 + x**2)
def N(b,d,lam):
    return (a + b - d) / lam
def R():
    return x+y
def d():
    return x+y
def b(BB):
    return np.sqrt((h-BB)**2 + y**2)

def l_eff(BB,k):
    return 20*np.log10((np.sqrt(2*np.pi*N(b(BB),d(),lamb[k]))/np.sqrt(np.tanh(2*np.pi*N(b(BB),d(),lamb[0])))))+5

def l_ri(delta):
    return Q+10*np.log10(F)-10*np.log10(omega)-20*np.log10(R())-0*delta*R()/1000
```

```python
plt.figure(dpi=150)
for k in range(7):
    res1=[]
    arg=[]
    arg = np.linspace(B, B1, 2000)
    for i in arg:
        res1.append(l_eff(i,k))
    plt.title(' ')
    plt.plot(arg, res1, label='.  = {}'.format(freq[k]))
plt.legend(loc=0, bbox_to_anchor=(1.5, 1))
plt.xlabel('B, ')
plt.ylabel('e, ')
plt.xlim(3, 0)
plt.show()

plt.figure(dpi=150)
for k in range(7):
    res1=[]
    arg=[]
    res2=[]
    arg = np.linspace(B, B1,2000)
    for i in arg:
        res1.append(l_eff(i,k))
        if l_ri(delta[k])>=0:
            #arg.append(i)
            res2.append(l_ri(delta[k]))
    plt.title('  ')
    plt.plot(arg, res2, label='.  = {}'.format(freq[k]))
plt.legend(loc=0, bbox_to_anchor=(1.5, 1))
plt.xlabel('B, ')
plt.ylabel('e, ')
plt.xlim(3, 0)
plt.show()
```
